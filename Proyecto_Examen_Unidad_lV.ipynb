# **Introduccion**

El rendimiento académico de los estudiantes, depende de muchos factores. Estos factores van desde el apoyo que reciben en casa, hasta el acceso a herramientas y recursos que les ayudan a aprender. En este estudio, se pretende analizar varias de estas variables, como la participación de los padres, el acceso a recursos educativos, la motivación de los estudiantes, y otros factores sociales y económicos.

El propósito principal de este análisis es entender cómo estas variables influyen en los resultados de los exámenes de los estudiantes. Para hacer esto, se utilizan dos herramientas estadísticas importantes: **ANOVA** y **prueba de Tukey**. **ANOVA** se usa para ver si hay diferencias significativas en los resultados de los exámenes de los estudiantes, dependiendo de factores como el nivel de participación de los padres o el acceso a recursos. La **prueba de Tukey** se usa después de ANOVA para averiguar específicamente cuáles de estos grupos tienen diferencias significativas.

Entre las variables que se estudian se encuentran el nivel educativo de los padres, si la escuela es pública o privada, la influencia de los amigos, y el acceso a internet y otros recursos educativos. Estos factores no solo afectan el rendimiento de los estudiantes de forma independiente, sino que pueden interactuar entre si, creando situaciones complejas. Esta investigación busca entender mejor cómo estos factores afectan en la educaión y cómo se pueden aplicar cambios para mejorar el rendimiento de los estudiantes, especialmente aquellos que enfrentan más desafíos.

Este estudio tiene como objetivo ayudar a identificar formas de apoyar a los estudiantes, para que todos tengan las mismas oportunidades de tener éxito en sus estudios.

Dentro de esta investigación tendremos 2 pruebas para realizar a cada una de las variables, una de ellas es: El Análisis de Varianza (ANOVA),esta es una herramienta estadística comúnmente utilizada para investigar el impacto de factores categóricos en variables numéricas (como los puntajes en los exámenes). ANOVA permite comparar las medias de tres o más grupos para ver si existen diferencias significativas entre ellos. Esta prueba se basa en la comparación de la varianza dentro de los grupos con la varianza entre los grupos, lo que ayuda a determinar si las diferencias observadas son lo suficientemente grandes como para no ser el resultado del azar.

Por otro lado, la prueba de Tukey, también conocida como prueba de comparación de medias de Tukey o HSD de Tukey (Honestly Significant Difference), es un método estadístico utilizado en el análisis de varianza (ANOVA) para realizar comparaciones múltiples entre las medias de diferentes grupos. Su objetivo principal es identificar cuáles pares de medias difieren significativamente, controlando la tasa de error tipo I (falsos positivos). Es especialmente útil cuando se tienen más de dos grupos y se desea determinar qué combinaciones presentan diferencias significativas.

Estas herramientas son útiles dentro de esta investigación ya que dentro de la educación se utiliza para comprender cómo diferentes factores, podrían afectar el rendimiento académico de los estudiantes, permitiendo a los investigadores tomar decisiones informadas para mejorar la calidad educativa.



# **Metodología:**

**Datos Utilizados:**
El análisis se basó en un conjunto de datos que incluye información sobre horas de estudio, asistencia, actividades extracurriculares, nivel de motivación, acceso a recursos, calidad del profesorado y otros factores relacionados con estudiantes. También se consideraron los puntajes en un examen como medida del rendimiento académico.

**Análisis Estadísticos Utilizados:**
ANOVA: Usado para determinar si existen diferencias significativas entre los grupos de cada variable categórica (como bajo, medio, elevado).
Prueba de Tukey: Realizada para comparar los promedios entre pares de grupos y explorar las diferencias más detalladas.

**Variables a evaluar:**
Incluyen el nivel educativo de los padres, el tipo de escuela, la influencia de los compañeros, el género, entre otras. Cada variable se analizó individualmente para determinar su relación con los puntajes de los exámenes y se utiliza la prueba de ANOVA y de Tukey para evaluar si existen una diferencia significativa entre cada nivel de medición.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Yarentzi34/Unidad-4/refs/heads/main/dataset.csv')
df.drop(columns=[], inplace=True)
df.dropna(inplace=True)
df
```

# **1. Participación Parental**

```python
df['participacion_parental'].unique()
```

Se utiliza este código para observar cuales son aquellas variables de respuesta que están presentes en la columna del DataFrame de la variable que queremos estudiar, en este caso la participacion parental.

Las variables de respuesta en este caso fueron bajo, medio y elevado.

*ANOVA*

```python
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
    df[df['participacion_parental'] == 'medio']['puntaje_examen'],
    df[df['participacion_parental'] == 'elevado']['puntaje_examen'],
)
# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadístico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusión: Hay diferencias significativas entre al menos dos grupos.")
else:
    print("Conclusión: No hay diferencias significativas entre los tratamientos.")
```

Los resultados del gráfico de ANOVA son: F-Estadístico = 80.46, Valor p < 0.00001 y un nivel de significancia del 0.05.

Como el valor p es menor que el nivel de significancia, indica que el nivel de participación parental muestra diferencias significativas en el rendimiento académico.

*TUKEY*

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```

En el análisis de este gráfico de la variable de la participación parental se muestra que todos los pares muestran diferencias significativas, y la variable de respuesta que más destaca dentro de las medias es el grupo de "elevada".

Se utiliza este código para observar cuales son aquellas variables de respuesta que están presentes en la columna del DataFrame de la variable que queremos estudiar, en esta caso los niveles de medición del tipo de escuela.

Las variables de respuesta de esta variable son: públicas o privadas.

*ANOVA*

```python
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['tipo_escuela'] == 'publico']['puntaje_examen'],
    df[df['tipo_escuela'] == 'privado']['puntaje_examen'],
)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadístico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusión: Hay diferencias significativas entre al menos dos tratamientos.")
else:
    print("Conclusión: No hay diferencias significativas entre los tratamientos.")
```

Los resultados del gráfico de ANOVA son: F-Estadístico = 0.75, Valor p = 0.38550 y un nivel de significancia del 0.05.

Como el valor p es mayor que el nivel de significancia, indica que el tipo de escuela no muestra diferencias significativas en el rendimiento académico.

En este caso, no se utiliza el gráfico de Tukey debido a que la variable del tipo de escuela muestra que no hay diferencias significativas entre los grupos y ambos presentan medias similares.

# **9. Influencia de compañeros**

```python
df['influencia_companieros'].unique()
```

Se utiliza este código para observar cuales son aquellas variables de respuesta que están presentes en la columna del DataFrame de la variable que queremos estudiar, en esta caso los niveles de medición de la influencia de los compañeros.

Las variables de respuesta de esta variable son:positivo, negativo o neutral.

*ANOVA*

```python
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['influencia_companieros'] == 'positivo']['puntaje_examen'],
    df[df['influencia_companieros'] == 'negativo']['puntaje_examen'],
    df[df['influencia_companieros'] == 'neutral']['puntaje_examen'],
)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadístico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusión: Hay diferencias significativas entre al menos dos tratamientos.")
else:
    print("Conclusión: No hay diferencias significativas entre los tratamientos.")
```

Los resultados del gráfico de ANOVA son: F-Estadístico = 32.18, Valor p < 0.00001 y un nivel de significancia del 0.05.

Como el valor p es menor que el nivel de significancia, indica que la influencia de los compañeros muestra diferencias significativas en el rendimiento académico.



*TUKEY*

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['influencia_companieros'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```

En el análisis de este gráfico de la variable de la influencia de compañeros se muestra que todos los pares muestran diferencias significativas, y la variable de respuesta que más destaca dentro de las medias es el grupo de "positivo".
