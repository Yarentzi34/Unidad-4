Dentro de esta investigacin tendremos 2 pruebas para realizar a cada una de las variables, una de ellas es: El Anlisis de Varianza (ANOVA),esta es una herramienta estadstica comnmente utilizada para investigar el impacto de factores categricos en variables numricas (como los puntajes en los exmenes). ANOVA permite comparar las medias de tres o ms grupos para ver si existen diferencias significativas entre ellos. Esta prueba se basa en la comparacin de la varianza dentro de los grupos con la varianza entre los grupos, lo que ayuda a determinar si las diferencias observadas son lo suficientemente grandes como para no ser el resultado del azar.

Por otro lado, la prueba de Tukey, tambin conocida como prueba de comparacin de medias de Tukey o HSD de Tukey (Honestly Significant Difference), es un mtodo estadstico utilizado en el anlisis de varianza (ANOVA) para realizar comparaciones mltiples entre las medias de diferentes grupos. Su objetivo principal es identificar cules pares de medias difieren significativamente, controlando la tasa de error tipo I (falsos positivos). Es especialmente til cuando se tienen ms de dos grupos y se desea determinar qu combinaciones presentan diferencias significativas.

Estas herramientas son tiles dentro de esta investigacin ya que dentro de la educacin se utiliza para comprender cmo diferentes factores, podran afectar el rendimiento acadmico de los estudiantes, permitiendo a los investigadores tomar decisiones informadas para mejorar la calidad educativa.

# **1. Participacin Parental**

```python
df['participacion_parental'].unique()
```

Se utiliza este cdigo para observar cuales son aquellas variables de respuesta que estn presentes en la columna del DataFrame de la variable que queremos estudiar, en este caso la participacion parental.

Las variables de respuesta en este caso fueron bajo, medio y elevado.

*ANOVA*

```python
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
    df[df['participacion_parental'] == 'medio']['puntaje_examen'],
    df[df['participacion_parental'] == 'elevado']['puntaje_examen'],
)
# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadstico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusin: Hay diferencias significativas entre al menos dos grupos.")
else:
    print("Conclusin: No hay diferencias significativas entre los tratamientos.")
```

Los resultados del grfico de ANOVA son: F-Estadstico = 80.46, Valor p < 0.00001 y un nivel de significancia del 0.05.

Como el valor p es menor que el nivel de significancia, indica que el nivel de participacin parental muestra diferencias significativas en el rendimiento acadmico.

*TUKEY*

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Grfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Mltiples comparaciones entre todos los pares (Tukey)")
```

En el anlisis de este grfico de la variable de la participacin parental se muestra que todos los pares muestran diferencias significativas, y la variable de respuesta que ms destaca dentro de las medias es el grupo de "elevada".

```python
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['tipo_escuela'] == 'publico']['puntaje_examen'],
    df[df['tipo_escuela'] == 'privado']['puntaje_examen'],
)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadstico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusin: Hay diferencias significativas entre al menos dos tratamientos.")
else:
    print("Conclusin: No hay diferencias significativas entre los tratamientos.")
```

En el anlisis de este grfico de la variable de la influencia de compaeros se muestra que todos los pares muestran diferencias significativas, y la variable de respuesta que ms destaca dentro de las medias es el grupo de "positivo".
